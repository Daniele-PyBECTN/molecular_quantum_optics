```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from scipy.signal import find_peaks, peak_widths

from qutip import *


font = {
    'family': 'serif',
    'color': 'black',
    'weight': 'normal',
    'size': 16,
}
```

```python
#SIMULATION RESONANT SATURATION
N_int = 50
vI = np.logspace(-1, 5, N_int)  # vector intensity in W/cm^2

Isat = 4e0    # W/(cm^2)
#Om_s = 2e-2     # GHz (frequency, already without 2/pi)

gam = .07    # radiative rate in GHz
Om_s = gam

Id = 5e2
scaling_exp_ds = 1

gam_lin = gam * (1.+ (.5*vI/Id)**2 )

gam_sim = gam * (1.+ (vI/Id)**scaling_exp_ds )
S_sim = (Om_s**2.*vI/Isat)/( gam_sim )**2

DW_sim = S_sim/vI * gam_sim**2
DW_sim = DW_sim/np.amax(DW_sim)

Fluo_sim = .5* S_sim/(1.+S_sim)
```

```python
#SIMULATION OFF RESONANT SATURATION

N_int_off = 100
vI_off = np.logspace(3, 7, N_int_off)  # vector intensity in W/cm^2

gam = .1
Om_s = gam

fact_rees = 10.

cf = 1
Isat = 4e1 * cf
Id = 1e4 * cf

gam_v = 15
eta = .3
scaling_exp_ds = 1


arr_pop_off = np.zeros(N_int_off)

vS = np.zeros(N_int_off)

eta_eff = 100.

Nv = 6
s_ = tensor(sigmam(), qeye(Nv))
b = tensor(qeye(2), destroy(Nv))
Delta = 0

for nI, Ioff in enumerate(vI_off):
    Om = Om_s * (Ioff/Isat)**.5

    Gam_p1 = eta_eff * eta**2 * gam * (Ioff/Id)**(1.*scaling_exp_ds)
    gam_resc = gam*(1.+(Ioff/Id)**(1.*scaling_exp_ds))

    #Gam_pump = eta**2 * Om**2 / (gam_v**2 + Gam_p1**2 )**.5
    Gam_pump = 4* (eta*Om)**2 /gam_v/ (1. + Gam_p1/gam_v )**2
    
    arr_pop_off[nI] = Gam_pump/(gam_resc)/(1.+Gam_pump/(gam_resc))
    
    c_jumps = [ (gam_resc)**.5 *s_, gam_v**.5 *b, (Gam_p1)**.5 *b*s_  ]
    #H = Delta * s_.dag()*s_ - Delta*b.dag()*b + Om* (k)**.5 * ( b*s_ + (b*s_).dag() )
    H = Delta * s_.dag()*s_ - 0*b.dag()*b + eta*Om * ( b*s_ + (b*s_).dag() )
    
    rhoss = steadystate(H, c_jumps)
    pop = expect(s_.dag()*s_, rhoss)
    vS[nI] = pop/(1.-pop)


S_off_sim = arr_pop_off / (1.-arr_pop_off) 

vpop = vS/(1+vS)

print(np.amax(arr_pop_off))
```
